Image Standization for Transfer learning

```python
input_path="C:\\Akash_damerla\\Epics\\resized_dataset\\Leaf_Curl\\"
output_path="C:\\Akash_damerla\\Epics\\Train_data\\Leaf_Curl\\"
```

```python
def augment_image(img):
    transformations = [
        lambda x: x,
        ImageOps.mirror,
        ImageOps.flip,
        lambda x: x.rotate(random.choice([90,180,270])),
        lambda x: ImageEnhance.Contrast(x).enhance(random.uniform(0.5,1.5)),
        lambda x: ImageEnhance.Color(x).enhance(random.uniform(0.7,1.3)),
        lambda x: ImageEnhance.Brightness(x).enhance(random.uniform(0.7,1.3))
    ]
    augmented = []
    for transform in transformations:
        augmented.append(transform(img))
    return augmented

def create_augmented_dataset(input_dir, output_dir, total_images=1000):
    os.makedirs(output_dir, exist_ok=True)
    images = [Image.open(os.path.join(input_dir, fname)) for fname in os.listdir(input_dir)]
    
    count = 0
    while count < total_images:
        img = random.choice(images)
        for aug_img in augment_image(img):
            if count >= total_images:
                break
            aug_img.save(os.path.join(output_dir, f"img_{count}.png"))
            count += 1

# Example usage:
create_augmented_dataset(input_path, output_path, total_images=1000)
```
